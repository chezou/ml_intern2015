# Cookpadのレシピから作ったword2vecモデル

gensimを使ってword2vecのモデルを扱います。

- Document: https://radimrehurek.com/gensim/models/word2vec.html 
- 関連記事: http://techlife.cookpad.com/entry/2015/02/27/093000

```python
%matplotlib inline

from os.path import expanduser
from visword2vec import visWord2Vec

vw2v = visWord2Vec(expanduser("~/data/recipe_steps.bin"))
vw2v_wiki = visWord2Vec(expanduser("~/data/wikipedia.bin"))
```

```python
# "チョコ"に類似している語
print("Cookpad")
vw2v.search(positive = [u'チョコ'])

print("\nWikipedia")
vw2v_wiki.search(positive = [u'チョコ'])
```

```python
# 赤味噌 - 名古屋 + 京都
print("Cookpad")
vw2v.search(positive = [u'赤味噌', u'京都'], negative=[u'名古屋'])

print("\nWikipedia")
vw2v_wiki.search(positive = [u'赤味噌', u'京都'], negative=[u'名古屋'])
```

```python
# "チョコ"に近い単語を2次元で可視化

print("Cookpad")
vw2v.plot(positive = [u'チョコ'])

print("Wikipedia")
vw2v_wiki.plot(positive = [u'チョコ'])
```
